### Month to date (MTD)  Total Loan Applications

### Good Loan Metrics

```python
good_loan = df[df['loan_status'].isin(['Fully Paid', 'Current'])]              #Good Loan

total_loan_application =  df['id'].count()

good_loan_application = good_loan['id'].count()
good_loan_amount_funded = good_loan['loan_amount'].sum()
good_loan_amount_received = good_loan['total_payment'].sum()

good_loan_amount_funded_millions = good_loan['loan_amount'].sum()/1000000
good_loan_amount_received_millions = good_loan['total_payment'].sum()/1000000

good_loan_rate = (good_loan_application/total_loan_application)*100

print("Good Loan Application:", good_loan_application)
print("Good Loan Funded amount: ${:.2f}M". format(good_loan_amount_funded_millions))
print("Good Loan Received Amount: ${:.2f}M". format(good_loan_amount_received_millions))
print("Good loan rate: {:.2f}%".format(good_loan_rate))
```

### Bad Loan Metrics

```python
bad_loan = df[df['loan_status'].isin(['Charged Off'])]

total_loan_application =  df['id'].count()

bad_loan_application = bad_loan['id'].count()
bad_loan_amount_funded = bad_loan['loan_amount'].sum()/1000000
bad_loan_amount_received = bad_loan['total_payment'].sum()/1000000

bad_loan_rate = (bad_loan_application/total_loan_application)*100

print("Good Loan Application:", bad_loan_application)
print("Good Loan Funded amount: ${:.2f}M". format(bad_loan_amount_funded))
print("Good Loan Received Amount: ${:.2f}M". format(bad_loan_amount_received))
print("Good loan rate: {:.2f}%".format(bad_loan_rate))
```

### Total Funded amount monthly trend

```python
monthly_funded = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort=False)['loan_amount']
      .sum()
      .div(1000000)
      .reset_index(name='loan_amount_millions')
)

plt.figure(figsize=(10, 5))
plt.fill_between(monthly_funded['month_name'], monthly_funded['loan_amount_millions'], 
                 color='yellow', alpha=0.5)
plt.plot(monthly_funded['month_name'], monthly_funded['loan_amount_millions'], 
         color='black', linewidth=2)

for i, row in monthly_funded.iterrows():
    plt.text(i, row['loan_amount_millions'] + 0.1, f"{row['loan_amount_millions']:.2f}", 
             ha='center', va='bottom', fontsize=9, rotation=0, color='black')

plt.title('Total Funded Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Funded Amount ($ Millions)')
plt.xticks(ticks=range(len(monthly_funded)), labels=monthly_funded['month_name'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Total Received Amount monthly trend

```python
monthly_funded = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort=False)['total_payment']
      .sum()
      .div(1000000)
      .reset_index(name='received_amount_millions')
)

plt.figure(figsize=(10, 5))
plt.fill_between(monthly_funded['month_name'], monthly_funded['received_amount_millions'], 
                 color='violet', alpha=0.5)
plt.plot(monthly_funded['month_name'], monthly_funded['received_amount_millions'], 
         color='black', linewidth=2)

for i, row in monthly_funded.iterrows():
    plt.text(i, row['received_amount_millions'] + 0.1, f"{row['received_amount_millions']:.2f}", 
             ha='center', va='bottom', fontsize=9, rotation=0, color='black')

plt.title('Total Received Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Received Amount ($ Millions)')
plt.xticks(ticks=range(len(monthly_funded)), labels=monthly_funded['month_name'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Loan Application trend by month

```python
monthly_applications = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort=False)['id']
      .count()
      .reset_index(name='loan_applications_count')
)

plt.figure(figsize=(10, 5))
plt.fill_between(monthly_applications['month_name'], monthly_applications['loan_applications_count'], 
                 color='red', alpha=0.5)
plt.plot(monthly_applications['month_name'], monthly_applications['loan_applications_count'], 
         color='darkblue', linewidth=2)

for i, row in monthly_applications.iterrows():
    plt.text(i, row['loan_applications_count'] + 0.5, f"{row['loan_applications_count']}", 
             ha='center', va='bottom', fontsize=9, rotation=0, color='black')

plt.title('Total Loan Applications by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Number of Applications')
plt.xticks(ticks=range(len(monthly_applications)), labels=monthly_applications['month_name'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Total funded amount by each state

```python
state_funding = df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=True)
state_funding_thousands = state_funding / 1000

plt.figure(figsize=(10, 8))
bars = plt.barh(state_funding_thousands.index, state_funding_thousands.values, color='darkblue')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 10, bar.get_y() + bar.get_height() / 2,
             f'{width:,.0f}k', va='center', fontsize=9)

plt.title('Total Funded Amount by State (in â‚¹ Thousands)')
plt.xlabel('Funded Amount (â‚¹ \'000)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

### Total received amount by each state

```python
state_funding = df.groupby('address_state')['total_payment'].sum().sort_values(ascending=True)
state_funding_thousands = state_funding / 1000

plt.figure(figsize=(10, 8))
bars = plt.barh(state_funding_thousands.index, state_funding_thousands.values, color='lightgreen')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 10, bar.get_y() + bar.get_height() / 2,
             f'{width:,.0f}k', va='center', fontsize=9)

plt.title('Total Received Amount by State (in â‚¹ Thousands)')
plt.xlabel('Funded Amount (â‚¹ \'000)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

### Funded amount by Loan Term

```python
term_funding_millions = df.groupby('term')['loan_amount'].sum() / 1000000

plt.figure(figsize=(5, 5))
plt.pie(
    term_funding_millions,
    labels=term_funding_millions.index,
    autopct=lambda p: f"{p:.1f}%\n${p*sum(term_funding_millions)/100:.1f}M",
    startangle=90,
    wedgeprops={'width': 0.6}
)

plt.gca().add_artist(plt.Circle((0, 0), 0.30, color='white'))
plt.title("Total Funded Amount by Term (in $ Millions)")
plt.show()
```

### Loan Purpose

```python
purpose_funding_millions = (df.groupby('purpose')['loan_amount'].sum().sort_values() / 1000000)

plt.figure(figsize=(10, 6))
bars = plt.barh(purpose_funding_millions.index, purpose_funding_millions.values, color='brown')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.1, bar.get_y() + bar.get_height() / 2,
             f'{width:.2f}M', va='center', fontsize=9)

plt.title('Total Funded Amount by Loan Purpose (â‚¹ Millions)', fontsize=14)
plt.xlabel('Funded Amount (â‚¹ Millions)')
plt.ylabel('Loan Purpose')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
